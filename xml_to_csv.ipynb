```python
import pandas as pd
import xml.etree.ElementTree as ET
from datetime import datetime
```

```python
def extract_XML(filename):
    FAN_lst = []
    XPN_lst = []
    PA_lst = []
    ETI_lst = []
    EAB_lst = []
    PRDF_lst = []
    APD_lst = []
    PDG_lst = []
    ACT_EED_lst = []
    KEYC_lst = []
    TECD_lst = []
    CPCM_lst = []
    STATUS_lst = []
    STATE_lst = []
    NSCTG_lst = []

    tree = ET.parse(filename+'.xml')
    root = tree.getroot()

    for patent_doc in root.findall('DOCUMENT'):
        FAN_tmp = None
        XPN_tmp = None
        PA_tmp = None
        ETI_tmp = None
        EAB_tmp = None
        PRDF_tmp = None
        APD_tmp = None
        PDG_tmp = None
        KEYC_tmp = None
        TECD_tmp = None
        CPCM_tmp = None
        STATUS_tmp = None
        STATE_tmp = None
        NSCTG_tmp = None
        ACT_EED_tmp = None
        PDA_tmp = None
        
    
        XPN_tmp = patent_doc.find('PRDF').get('original')
        
        FAN_tmp = patent_doc.find('FAN').text
        
        PRDF_tmp = patent_doc.find('PRDF').text
        
        APD_tmp = patent_doc.find('APD').text if patent_doc.find('APD') is not None else None
        
        PDG_tmp = patent_doc.find('PDG').text if patent_doc.find('PDG') is not None else None
        
        ACT_EED_tmp = patent_doc.find('ACT.EED').text
        
        ETI_tmp = patent_doc.find('ETI').text
        
        EAB_tmp = patent_doc.find('EAB').text if patent_doc.find('EAB') is not None else None
        
        PA_XML_LIST = patent_doc.find('PA-LIST')
        PA_tmp = [assignee.text for assignee in PA_XML_LIST.findall('PA')]
        
        KEYC_XML_LIST = patent_doc.find('KEYC-LIST')
        KEYC_tmp = [keyword.text for keyword in KEYC_XML_LIST.findall('KEYC')]
        
        TECD_XML_LIST = patent_doc.find('TECD-LIST')
        TECD_tmp = [domain.text for domain in TECD_XML_LIST.findall('TECD')] if TECD_XML_LIST is not None else None
        
        STATUS_tmp = patent_doc.find('ACT.STATUS').text
        
        STATE_tmp = patent_doc.find('ACT.STATE').text
        
        CPCM_tmp = patent_doc.find('CPCM').text if patent_doc.find('CPCM') is not None else None
        
        
        
        
        

        FAN_lst.append(FAN_tmp)        
        XPN_lst.append(XPN_tmp)
        ETI_lst.append(ETI_tmp)
        EAB_lst.append(EAB_tmp)
        PA_lst.append(' | '.join(PA_tmp) if PA_tmp is not None else PA_tmp)
        PRDF_lst.append(PRDF_tmp)
        APD_lst.append(APD_tmp)
        PDG_lst.append(PDG_tmp)
        KEYC_lst.append(' | '.join(KEYC_tmp) if KEYC_tmp is not None else KEYC_tmp)
        CPCM_lst.append(CPCM_tmp)
        TECD_lst.append(' | '.join(TECD_tmp) if TECD_tmp is not None else TECD_tmp)
        STATUS_lst.append(STATUS_tmp)
        STATE_lst.append(STATE_tmp)
        NSCTG_lst.append(NSCTG_tmp)


    data = {'Questel Family ID':FAN_lst, 'Publication No':XPN_lst, 'Title':ETI_lst, 'Abstract': EAB_lst, 'Current Assignee': PA_lst, 
            'Earliest Priority Date': PRDF_lst, 'Application Date': APD_lst, 'Grant Date': PDG_lst,
            'Keywords':KEYC_lst, 'Technology Domain':TECD_lst, 'Main CPC':CPCM_lst,
            'Status': STATUS_lst, "State": STATE_lst, "Non-self citing patent count": NSCTG_lst}
    df = pd.DataFrame(data)
    return df
```

```python
def extract_tech_domain(filename):
    XPN_lst = []
    TECD_lst = []

    tree = ET.parse(filename+'.xml')
    root = tree.getroot()

    for patent_doc in root.findall('QOdocument'):

        XPN_tmp = None
        TECD_tmp = None

        for field in patent_doc.findall('QOfield'):
            match field.attrib.get('name'):

                case 'XPN':
                    par = field.find('QOpar')
                    XPN_tmp = par.find('QOsen').text

                case 'TECD':
                    par = field.find('QOpar')
                    for tech in par.findall('QOsen'):
                        TECD_tmp = tech.text
                        
                        XPN_lst.append(XPN_tmp)
                        TECD_lst.append(TECD_tmp)
                        




    data = {'Publication No':XPN_lst, 'Technology Domain':TECD_lst}
    df = pd.DataFrame(data)
    return df
```

```python
def extract_assignee(filename):
    XPN_lst = []
    PA_lst = []

    tree = ET.parse(filename+'.xml')
    root = tree.getroot()

    for patent_doc in root.findall('QOdocument'):

        XPN_tmp = None
        PA_tmp = None

        for field in patent_doc.findall('QOfield'):
            match field.attrib.get('name'):

                case 'XPN':
                    par = field.find('QOpar')
                    XPN_tmp = par.find('QOsen').text

                case 'PA':
                    par = field.find('QOpar')
                    for assignee in par.findall('QOsen'):
                        PA_tmp = assignee.text
                        
                        XPN_lst.append(XPN_tmp)
                        PA_lst.append(PA_tmp)
                        




    data = {'Publication No':XPN_lst, 'Current Assignee':PA_lst}
    df = pd.DataFrame(data)
    return df
```
